```python
# Tarea 2 - RDD con datos desde API pública

API utilizada: REST Countries
Objetivo:
Obtener datos desde una API pública, convertirlos a RDD en PySpark y realizar estadísticas descriptivas.
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.master("local[*]").getOrCreate()
sc = spark.sparkContext
```

```python
import requests
import json

url = "https://restcountries.com/v3.1/all?fields=name,population,region"
response = requests.get(url)

print("Status code:", response.status_code)

data = response.json()

type(data), type(data[0])
```

```python
rdd_paises = sc.parallelize(data)

rdd_paises.take(1)
```

```python
rdd_poblacion = rdd_paises.map(lambda x: x.get("population", 0))

rdd_poblacion.take(5)
```

```python
print("Número de países:", rdd_poblacion.count())
print("Promedio población:", rdd_poblacion.mean())
print("Población mínima:", rdd_poblacion.min())
print("Población máxima:", rdd_poblacion.max())
print("Desviación estándar:", rdd_poblacion.stdev())
```
