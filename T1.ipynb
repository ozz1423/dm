```python
# Análisis de Datos de Salud Mental 2024–2025

**Herramienta:** PySpark  
**Entorno:** WSL2 + Ubuntu  

## Conjunto de datos
Conjunto de datos que muestra las consultas de salud mental otorgadas en las unidades médicas de los Servicios de Salud de Nuevo León generadas durante el periodo de octubre 2024 hasta agosto 2025.

## Objetivo
Realizar análisis exploratorio y descriptivo sobre los registros de salud mental 2024–2025 utilizando PySpark.
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.master("local[*]").getOrCreate()
```

```python
df = spark.read.csv(
    "2024_2025_salud_mental.csv",
    header=True,
    inferSchema=True
)

df.show(5)
```

```python
df.printSchema()
df.count()
```

```python
from pyspark.sql.functions import to_date, when, col

df = df.withColumn("fecha", to_date("fecha", "dd/MM/yyyy"))

df = df.withColumn(
    "peso",
    when(col("peso") == "sin valor", None).otherwise(col("peso")).cast("double")
)

df = df.withColumn(
    "altura",
    when(col("altura") == "sin valor", None).otherwise(col("altura")).cast("double")
)

df.printSchema()
```

```python
df.select("edad", "peso", "altura").describe().show()
```

```python
df = df.withColumn(
    "grupo_edad",
    when(col("edad") < 18, "Menor")
    .when((col("edad") >= 18) & (col("edad") < 60), "Adulto")
    .otherwise("Adulto Mayor")
)

df.groupBy("grupo_edad").count().show()
```
